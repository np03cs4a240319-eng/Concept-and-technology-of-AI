```python
import numpy as np

# 1. Create an array with values ranging from 10 to 49
arr_range = np.arange(10, 50)
print("1. Array 10 to 49:\n", arr_range)

# 2. Create a 3x3 matrix with values from 0 to 8
arr_matrix = np.arange(9).reshape(3, 3)
print("\n2. 3x3 Matrix 0 to 8:\n", arr_matrix)

# 3. Create a 3x3 identity matrix
identity = np.eye(3)
print("\n3. Identity Matrix:\n", identity)

# 4. Create a random array of size 30 and find the mean
rand_arr = np.random.random(30)
print("\n4. Mean of random 30 values:", rand_arr.mean())

# 5. Create a 10x10 random array and find min & max
rand_10x10 = np.random.random((10, 10))
print("\n5. Min:", rand_10x10.min(), " Max:", rand_10x10.max())

# 6. Create a zero array of size 10 and replace 5th element with 1
arr_zero = np.zeros(10)
arr_zero[4] = 1
print("\n6. Zero array with 5th element = 1:\n", arr_zero)

# 7. Reverse an array
arr = np.array([1, 2, 0, 0, 4, 0])
rev_arr = arr[::-1]
print("\n7. Reversed array:\n", rev_arr)

# 8. Create a 2D array with 1 on border and 0 inside
border = np.ones((5, 5))
border[1:-1, 1:-1] = 0
print("\n8. Border 1, inside 0:\n", border)

# 9. 8x8 checkerboard pattern
checkerboard = np.zeros((8, 8))
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n9. Checkerboard pattern:\n", checkerboard)
```

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
add_xy = x + y
print("1. Addition:\n", add_xy)

# 2. Subtract the two arrays
sub_xy = x - y
print("\n2. Subtraction:\n", sub_xy)

# 3. Multiply the array with an integer (example: multiply x by 3)
mul_x = x * 3
print("\n3. x multiplied by 3:\n", mul_x)

# 4. Square of each element of array x
square_x = x ** 2
print("\n4. Square of x:\n", square_x)

# 5. Dot products
dot_vw = np.dot(v, w)
print("\n5(a). Dot product v·w =", dot_vw)

dot_xv = np.dot(x, v)
print("\n5(b). Dot product x·v:\n", dot_xv)

dot_xy = np.dot(x, y)
print("\n5(c). Dot product x·y:\n", dot_xy)

# 6. Concatenate x and y along rows
concat_xy_row = np.concatenate((x, y), axis=0)
print("\n6(a). Concatenate x & y along rows:\n", concat_xy_row)

# Concatenate v and w along columns
concat_vw_col = np.column_stack((v, w))
print("\n6(b). Concatenate v & w along columns:\n", concat_vw_col)

# 7. Concatenate x and v (observe error and fix)
try:
    try_concat = np.concatenate((x, v), axis=0)
    print("\n7. Concatenate x & v:\n", try_concat)
except ValueError as e:
    print("\n7. Error while concatenating x & v:", e)

# Correct concatenation after reshaping v
correct_concat = np.concatenate((x, v.reshape(1, 2)), axis=0)
print("\nCorrect concatenation (after reshape):\n", correct_concat)
```

```python
import numpy as np

# 1. Prove A * A^-1 = I

A = np.array([[3, 4],
              [7, 8]])

print("Matrix A:")
print(A)

A_inv = np.linalg.inv(A)
print("\nInverse of A:")
print(A_inv)

identity_check = A.dot(A_inv)
print("\nA * A_inv (Should be Identity Matrix):")
print(identity_check)


# 2. Prove AB != BA

B = np.array([[5, 3],
              [2, 1]])

print("\nMatrix B:")
print(B)

AB = A.dot(B)
BA = B.dot(A)

print("\nAB =")
print(AB)

print("\nBA =")
print(BA)

print("\nIs AB equal to BA? ->", np.array_equal(AB, BA))


# 3. Prove (AB)^T = B^T A^T

left = (A.dot(B)).T
right = B.T.dot(A.T)

print("\n(AB)^T =")
print(left)

print("\nB^T A^T =")
print(right)

print("\nIs (AB)^T equal to B^T A^T? ->", np.array_equal(left, right))


# 4. Solve the system of linear equations
#
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9
#
# Using inverse method

A2 = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

B2 = np.array([-1, -3, 9])

print("\n\nCoefficient Matrix A2:")
print(A2)

print("\nConstant Matrix B2:")
print(B2)

A2_inv = np.linalg.inv(A2)
solution1 = A2_inv.dot(B2)

print("\nSolution using Inverse Method (A^-1 * B):")
print("x =", solution1[0])
print("y =", solution1[1])
print("z =", solution1[2])

# 5. Solve using np.linalg.solve (recommended method)

solution2 = np.linalg.solve(A2, B2)

print("\nSolution using numpy.linalg.solve:")
print("x =", solution2[0])
print("y =", solution2[1])
print("z =", solution2[2])
```
