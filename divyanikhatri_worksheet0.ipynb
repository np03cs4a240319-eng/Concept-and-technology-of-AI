```python
# Given study time data (example)
time_data = [1, 4, 6, 2, 3.5, 5, 7, 2.5, 4.2]

# 1. Create empty lists
low = []
moderate = []
high = []

# 2. Iterate and classify
for t in time_data:
    if t < 3:
        low.append(t)
    elif 3 <= t <= 5:
        moderate.append(t)
    else:
        high.append(t)

# 3. Print results
print("Low Study Time (< 3 hrs):", low)
print("Moderate Study Time (3 - 5 hrs):", moderate)
print("High Study Time (> 5 hrs):", high)
```

```python
# 1. Count days with low study time
low_count = len(low)
print("Number of days with low study time:", low_count)

# 2. Count days with moderate study time
moderate_count = len(moderate)
print("Number of days with moderate study time:", moderate_count)

# 3. Count days with high study time
high_count = len(high)
print("Number of days with high study time:", high_count)
```

```python
# Given study time data (in hours)
time_data = [1, 4, 6, 2, 3.5, 5, 7, 2.5, 4.2]

# 1 & 2. Convert hours to minutes and store in a new list
study_minutes = [t * 60 for t in time_data]

# 3. Print the converted values
print("Study time in minutes:", study_minutes)
```

```python
# Sample time_data: [study, entertainment, sleep] for each day
time_data = [
    [2, 3, 7],
    [4, 2, 6],
    [6, 1, 5],
    [3, 4, 6],
    [5, 2, 7]
]

# 1. Create empty lists
study_hours = []
entertainment_hours = []
sleep_hours = []

# 2. Extract values into each list
for record in time_data:
    study_hours.append(record[0])
    entertainment_hours.append(record[1])
    sleep_hours.append(record[2])

# 3. Calculate and print averages
avg_study = sum(study_hours) / len(study_hours)
avg_entertainment = sum(entertainment_hours) / len(entertainment_hours)
avg_sleep = sum(sleep_hours) / len(sleep_hours)

print("Average study hours:", avg_study)
print("Average entertainment hours:", avg_entertainment)
print("Average sleep hours:", avg_sleep)
```
