```python
# Given study time data (example)
time_data = [1, 4, 6, 2, 3.5, 5, 7, 2.5, 4.2]

# 1. Create empty lists
low = []
moderate = []
high = []

# 2. Iterate and classify
for t in time_data:
    if t < 3:
        low.append(t)
    elif 3 <= t <= 5:
        moderate.append(t)
    else:
        high.append(t)

# 3. Print results
print("Low Study Time (< 3 hrs):", low)
print("Moderate Study Time (3 - 5 hrs):", moderate)
print("High Study Time (> 5 hrs):", high)
```

```python
# 1. Count days with low study time
low_count = len(low)
print("Number of days with low study time:", low_count)

# 2. Count days with moderate study time
moderate_count = len(moderate)
print("Number of days with moderate study time:", moderate_count)

# 3. Count days with high study time
high_count = len(high)
print("Number of days with high study time:", high_count)
```

```python
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.

    Args:
        nested_list (list): A list that may contain integers or other nested lists of integers.

    Returns:
        int: The total sum of all integers in the nested list, including those in sublists.

    Example:
    >>> sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]])
    36
    >>> sum_nested_list([1, [2, 3], [4, [5]]])
    15
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if the element is a list
            total += sum_nested_list(element)  # Recursively sum the nested list
        else:
            total += element  # Add the number to the total
    return total

# Test the function
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total sum of nested list:", result)  # Output should be 36
```

```python
def generate_permutations(s):
    """
    Generate all unique permutations of a string recursively.

    Args:
        s (str): Input string.

    Returns:
        list: List of all unique permutations.
    """
    if len(s) == 1:
        return [s]  # Base case: single character string

    permutations = []
    for i, char in enumerate(s):
        # Exclude the current character and get permutations of the remaining string
        remaining = s[:i] + s[i+1:]
        for p in generate_permutations(remaining):
            permutations.append(char + p)

    # Remove duplicates (for strings with repeated characters)
    return list(set(permutations))

# Test the function
print(generate_permutations("abc"))
# Output: ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']

print(generate_permutations("aab"))
# Output: ['aab', 'aba', 'baa']
```

```python
# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

def calculate_directory_size(directory):
    """
    Recursively calculate total size of a directory including all nested files and subdirectories.

    Args:
        directory (dict): A dictionary representing files and subdirectories.

    Returns:
        int: Total size in KB.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            # If it's a subdirectory, recursively calculate its size
            total_size += calculate_directory_size(value)
        else:
            # If it's a file, add its size
            total_size += value
    return total_size
total_size = calculate_directory_size(directory_structure)
print("Total directory size:", total_size, "KB")  # Expected: 1400 KB
```

```python
def min_coins_with_combination(coins, amount):
    """
    Finds the minimum number of coins needed to make a given amount and returns the coins used.

    Parameters:
        coins (list of int): Available coin denominations.
        amount (int): Target amount.

    Returns:
        tuple: (min_coins_count, list_of_coins_used)
               Returns (-1, []) if it is impossible.
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    # To track coins used
    coin_used = [-1] * (amount + 1)

    for coin in coins:
        for i in range(coin, amount + 1):
            if dp[i - coin] + 1 < dp[i]:
                dp[i] = dp[i - coin] + 1
                coin_used[i] = coin

    if dp[amount] == float('inf'):
        return -1, []

    # Reconstruct coins used
    result_coins = []
    curr = amount
    while curr > 0:
        result_coins.append(coin_used[curr])
        curr -= coin_used[curr]

    return dp[amount], result_coins

# Test the function
coins = [1, 2, 5]
amount = 11

min_count, coins_used = min_coins_with_combination(coins, amount)
print("Minimum coins needed:", min_count)        # Output: 3
print("Coins used to make amount:", coins_used)  # Output: [5, 5, 1]
```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the Longest Common Subsequence (LCS) between two strings using DP.

    Args:
        s1 (str): First string.
        s2 (str): Second string.

    Returns:
        int: Length of the LCS.
    """
    m, n = len(s1), len(s2)
    # Create DP table with (m+1) x (n+1) initialized to 0
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

# Test the function
s1 = "abcde"
s2 = "ace"
lcs_length = longest_common_subsequence(s1, s2)
print("Length of LCS:", lcs_length)
```
