```python

import pandas as pd
import numpy as np

df = pd.read_csv('/content/Titanic-Dataset.csv')
#1.
df['fare_per_year'] = df['Fare'] / df['Age']
#2.
high_fare_age = df[df['fare_per_year'] >5 ]
#3.
high_fare_age_sort = high_fare_age.sort_values('fare_per_year', ascending=False)
#4.
result = high_fare_age_sort[['Name', 'fare_per_year']]
#5.
print(result.head())

print(f"Name: {result.iloc[0]['Name']}")
print(f"Fare per year: {result.iloc[0]['fare_per_year']:.2f}")
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
#1.
df['fare_per_class'] = df['Fare'] / df['Pclass']
#2.
adult_males = df[(df['Sex']=='male') & (df['Age'] >= 18)]
#3.
adult_males_sort = adult_males.sort_values('fare_per_class', ascending=False)
#4.
result = adult_males_sort[['Name', 'Age', 'fare_per_class']]
#5.
print(result.head())

print(f"Name: {result.iloc[0]['Name']}")
print(f"Age: {result.iloc[0]['Age']}")
print(f"Fare per class: {result.iloc[0]['fare_per_class']:.2f}")
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
#1.
total_fare = df['Fare'].sum()
#2.
first_class_fare = df[df['Pclass']==1]['Fare'].sum()
second_class_fare = df[df['Pclass']==2]['Fare'].sum()
third_class_fare = df[df['Pclass']==3]['Fare'].sum()

#3
class_fares = [first_class_fare, second_class_fare, third_class_fare]
#4.
first_class_pro = (first_class_fare / total_fare)*100
second_class_pro = (second_class_fare / total_fare)*100
third_class_pro = (third_class_fare / total_fare)*100

print(f"Total Fare Revenue: ${total_fare:.2f}")
print(f"\nBy Class:")
print(f"First Class:  ${first_class_fare:.2f} ({first_class_pro:.1f}%)")
print(f"Second Class: ${second_class_fare:.2f} ({second_class_pro:.1f}%)")
print(f"Third Class:  ${third_class_fare:.2f} ({third_class_pro:.1f}%)")
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
#1.
def get_age_group(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(get_age_group)

# 2
total_passengers = len(df)

for group in ['child', 'adult', 'senior', 'Unknown']:
    count = (df['age_group'] == group).sum()
    percentage = (count / total_passengers) * 100
    print(f"{group}: {count} passengers ({percentage:.1f}%)")
```
